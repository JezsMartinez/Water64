```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Medium (69 GRID)

cpbe69_m = np.genfromtxt('cpbe69s_m.txt')
npbe69_m = np.genfromtxt('npbe69s_m.txt')
ippbe_m = (cpbe69_m-npbe69_m)*27.211385 #Hartrees to eV

cmp269_m = np.genfromtxt('cmp269s_m.txt')
nmp269_m = np.genfromtxt('nmp269s_m.txt')
ipmp2_m = (cpbe69_m-npbe69_m)*27.211385 #Hartrees to eV


fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,4))

x = ippbe_m
y = ipmp2_m
sns.distplot(x, ax=ax1, bins=8, label=['PBE m'])
sns.distplot(y, ax=ax1, bins=8, label=['MP2 m'])
ax1.axvline(ippbe_m.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
ax1.axvline(ipmp2_m.mean(), color='k', linestyle='dashed', linewidth=1)
ax1.legend(loc='upper right')
ax1.set_xlim([7,15])
ax1.set_xlabel('IP/eV', fontsize=12)
print("IP average PBEm: ",np.average(ippbe_m))
print("IP average MP2m: ",np.average(ipmp2_m))

#Small (32 GRID)

cpbe69_s = np.genfromtxt('cpbe32s_s.txt')
npbe69_s = np.genfromtxt('npbe32s_s.txt')
ippbe_s = (cpbe69_s - npbe69_s)*27.211385 #Hartrees to eV

cmp269_s = np.genfromtxt('cmp232s_s.txt')
nmp269_s = np.genfromtxt('nmp232s_s.txt')
ipmp2_s = (cpbe69_s-npbe69_s)*27.211385 #Hartrees to eV

x = ippbe_s
y = ipmp2_s
sns.distplot(x, ax=ax2, bins=8, label=['MP2 s'])
sns.distplot(y, ax=ax2, bins=8, label=['PBE s'])
ax2.axvline(ippbe_s.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
ax2.axvline(ipmp2_s.mean(), color='k', linestyle='dashed', linewidth=1)
ax2.legend(loc='upper right')
ax2.set_xlim([7,15])
ax2.set_xlabel('IP/eV', fontsize=12)
print("IP average PBEs: ",np.average(ippbe_s))
print("IP average MP2s: ",np.average(ipmp2_s))
```
