```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#IP 68 and 69 spherical GRID

fig, ax = plt.subplots(2,2,figsize=(12,10))
cmp268t = np.genfromtxt('cmp268t.txt')
nmp268t = np.genfromtxt('nmp268t.txt')
ipmp268t = (cmp268t-nmp268t)*27.211385

x=ipmp268t
sns.distplot(x, ax=ax[0,0], bins=8, label=['MP2 T G68'])
#sns.regplot(x=x, y=y, color="g")
ax[0,0].axvline(ipmp268t.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
#ax[0,0].hist(x=ipmp268t, bins=8, label=['MP2 T G68'])
ax[0,0].legend(loc='upper right')
ax[0,0].set_xlim([10,14])
ax[0,0].set_xlabel('IP/eV', fontsize=12)
print("IP average T 68: ",np.average(ipmp268t))

cmp269t = np.genfromtxt('cmp269t.txt')
nmp269t = np.genfromtxt('nmp269t.txt')
ipmp269t = (cmp269t-nmp269t)*27.211385

x=ipmp269t
sns.distplot(x, ax=ax[0,1], bins=8, label=['MP2 T G69'])
ax[0,1].axvline(ipmp269t.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
#ax[0,1].hist(x=ipmp269t, bins=8, label=['MP2 T G69'])
ax[0,1].legend(loc='upper right')
ax[0,1].set_xlim([10,14])
ax[0,1].set_xlabel('IP/eV', fontsize=12)
print("IP average T 69: ",np.average(ipmp269t))

cmp268s = np.genfromtxt('cmp268s.txt')
nmp268s = np.genfromtxt('nmp268s.txt')
ipmp268s = (cmp268s-nmp268s)*27.211385

x=ipmp268s
sns.distplot(x, ax=ax[1,0], bins=8, label=['MP2 S G68'])
ax[1,0].axvline(ipmp268s.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
#ax[1,0].hist(x=ipmp268s, bins=8, label=['MP2 S G68'])
ax[1,0].legend(loc='upper right')
ax[1,0].set_xlim([7,16])
ax[1,0].set_xlabel('IP/eV', fontsize=12)
ax[1,0].set_xticks(np.arange(7, 16, 1))
print("IP average S 68: ",np.average(ipmp268s))

cmp269s = np.genfromtxt('cmp269s.txt')
nmp269s = np.genfromtxt('nmp269s.txt')
ipmp269s = (cmp269s-nmp269s)*27.211385

x=ipmp269s
sns.distplot(x, ax=ax[1,1], bins=8, label=['MP2 S G69'])
ax[1,1].axvline(ipmp269s.mean(), color='k', linestyle='dashed', linewidth=1, label='Average IP')
#ax[1,1].hist(x=ipmp269s, bins=8, label=['MP2 S G69'])
ax[1,1].legend(loc='upper right')
ax[1,1].set_xlim([7,16])
ax[1,1].set_xlabel('IP/eV', fontsize=12)
ax[1,1].set_xticks(np.arange(7, 16, 1))
print("IP average S 69: ",np.average(ipmp269s))
```

```python
## VBM
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,4))

x=ipmp268s
sns.kdeplot(x, ax=ax1, label=['MP2 S G69'])
point = sns.distplot(x, ax=ax1, bins=8, label=['MP2 S G69'])
data_x, data_y = point.lines[0].get_data()
print(data_x[22:42])
datax = data_x[20:42]
datay = data_y[20:42]
ax1.set_xlim([6,16])
sns.regplot(x=datax, y=datay, color="g", ax=ax1)

y=ipmp269s
points = sns.distplot(y, ax=ax2, bins=8, label=['PBE S G69'])
data_x, data_y = points.lines[0].get_data()
print(data_x[83:96])
datax = data_x[83:96]
datay = data_y[83:96]
ax2.set_xlim([6,16])
sns.regplot(x=datax, y=datay, color="g", ax=ax2)
```
